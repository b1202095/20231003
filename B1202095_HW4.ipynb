```
a = input("請輸入五個成績")
b = a.split(' ')
b.remove(max(b))
b.remove(min(b))
average = (int(b[0])+int(b[1])+int(b[2]))/3
print(average)
```

```
a = int(input('請輸入第一個數字:'))
b = int(input('請輸入第二個數字:'))
c = input('請輸入一個運算符號:')
if c == '+':
  result = a+b
elif c == '-':
  result = a-b
elif c == '/':
  result = a/b
else:
  result = a*b
print("計算結果=",str(result))
```
